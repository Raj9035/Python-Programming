```python
card_data = pd.read_csv('D:/Evarcity/R-Programming/card.txt', sep=",")
```

```python
# Step 1: check data type: if date column is not in date format change it to date foramt
# Step: Day, month, year, Time, Weekend, Holiday
```

```python
## Convert the date and time

card_data['application_date'] = pd.to_datetime(card_data['application_date'], errors='coerce')
```

```python
card_data['application_date']=pd.to_datetime(card_data.application_date, format='%m/%d/%Y %H:%M')
```

```python
## Split the data into year, month, day, hour, minute and seconds

card_data[['year','month','day','hour']] = card_data['application_date'].apply(lambda x:pd.Series({'year':x.year,'month':x.month,'day':x.day,'hour':x.hour}))
```

```python
card_data
```

```python
## Divide the data based on morning and evening time 
card_data['TimePeriod'] = pd.cut(card_data['hour'], bins=[-1,12,24], labels=['morning','evening'])
```
