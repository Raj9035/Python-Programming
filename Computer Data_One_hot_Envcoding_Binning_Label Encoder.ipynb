```python
customer_data = pd.read_csv('D:/Evarcity/R-Programming/Customer_Data.csv')
```

```python
customer_data.head(5)
```

```python
## Binning

customer_data['Age_category'] = pd.cut(customer_data['Age'],bins =[1,18,60,100], labels = ['Child','Adult','Senior'])
```

```python
## Use 'inf' if your data is updating daily (Binning), using numpy

customer_data['Age_category'] = pd.cut(customer_data['Age'],bins =[-np.inf,18,60,np.inf], labels = ['Child','Adult','Senior'])
```

```python
## Use 'inf' if your data is updating daily (Binning)

customer_data['Age_category'] = pd.cut(customer_data['Age'],bins =['-inf',18,60,'inf'], labels = ['Child','Adult','Senior'])
```
