```python
import pandas as pd
```

```python
import numpy as np
```

```python
titanic_data = pd.read_csv('D:/Evarcity/R-Programming/titanic.csv')
```

```python
titanic_data.head(5)
```

```python
## Without using any package performing scaling(normilaztion) if you dont have outliers

titanic_data['Normaliztion Fare'] = (titanic_data.Fare - titanic_data.Fare.min())/(titanic_data.Fare.max()-titanic_data.Fare.min())
```

```python
titanic_data
```

```python
## Standardization with out using function or package (if you have outliers)

titanic_data['Standardization Fare'] = (titanic_data.Fare - titanic_data.Fare.mean())/(titanic_data.Fare.std())
```

```python
titanic_data
```

```python
## Imputation: to find missing Values
titanic_data.isnull().any()
```

```python
## to get count of missing values

titanic_data.isnull().sum()
```

```python
## to get percentage of missing values

titanic_data.isnull().sum()/len(titanic_data)*100
```

```python
## replcaing missing value in AGe colum

m1 = titanic_data.Age.mean()

titanic_data['Age'] = titanic_data.Age.fillna(m1)
```

```python
## Relacing Embarked column with mode

titanic_data.Embarked.value_counts()
```

```python
titanic_data.Embarked.value_counts().idxmax()
```

```python
titanic_data['Embarked'] = titanic_data.Embarked.fillna('S')
```
