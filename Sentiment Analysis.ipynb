```python
import nltk
```

```python
nltk.download("vader_lexicon")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sent_analysis=SentimentIntensityAnalyzer()
```

```python
#get the first review

customer_review.iloc[0]['text']
```

```python
## Checking the ploarity score

sent_analysis.polarity_scores(customer_review.iloc[0]['text'])
```

```python
#calculate polarity score of all reviews

customer_review['polarity_score'] = customer_review['text'].apply(lambda x:sent_analysis.polarity_scores(x))
```

```python
customer_review['compound_score']= customer_review['polarity_score'].apply(lambda x:x['compound'])
```

```python
#classify feedback

import numpy as np

customer_review['Neg_pos'] = customer_review['compound_score'].apply(lambda x:np.where(x>0.05,'Positive','Negative'))
```

```python
#get the count of negative and positive feedbacks

customer_review["Neg_pos"].value_counts()
```
